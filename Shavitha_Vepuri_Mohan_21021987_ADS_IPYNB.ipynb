```python
import wbgapi as wb
```

```python
#Creating 1st data frame
#Climate change indicators
indicator_ids1 = ['EG.ELC.ACCS.ZS','ER.H2O.FWTL.ZS','AG.LND.FRST.ZS','AG.LND.IRIG.AG.ZS']
country_codes = ['USA','CAN','IND','JPN','CHN','THA','CRI','AUS','AFG','NZL']
my_dataframe1  = wb.data.DataFrame(indicator_ids1, country_codes, mrv=10)
print(my_dataframe1)
```

```python
#Creating 2nd data frame
#Aid Effectiveness indicators
indicator_ids1 = ['DT.ODA.ODAT.GI.ZS','SH.STA.ANVC.ZS','SH.STA.STNT.ZS','SE.PRM.CMPT.ZS']
country_codes = ['USA','CAN','IND','JPN','CHN','THA','CRI','AUS','AFG','NZL']
my_dataframe2  = wb.data.DataFrame(indicator_ids1, country_codes, mrv=10)
print(my_dataframe2)
```

```python
#Transforming the data
my_dataframe1.groupby('economy').transform(lambda x : x.mean())
```

```python
#pandas.apply()
my_dataframe2.groupby('economy').apply(lambda x: x.astype(float).max())
```

```python
#pandas.groupby()
x = my_dataframe1.groupby('economy')['YR2016'].mean()
y = my_dataframe1.groupby('economy')['YR2017'].median()
z = my_dataframe2.groupby('economy')['YR2016'].mean()
a = my_dataframe2.groupby('economy')['YR2017'].median()
print(x,y,z,a)
```

```python
#pandas.privottable()
import numpy as np
first = my_dataframe2.pivot_table(index=['series'],aggfunc={'YR2016':np.mean})
second = my_dataframe2.pivot_table(index=['series'],aggfunc={'YR2016':np.median})
print(first,second)
```

```python
#Filtering the data 
my_dataframe2.groupby('economy').filter(lambda x : x['YR2014'].median() >= 25)
```

```python
import matplotlib.pyplot as plt
a = my_dataframe1['YR2011']
B = my_dataframe1['YR2019']
plt.scatter(a, B, c ="blue")
# To show the plot
plt.show()
```
